### Branching using Conditional Statements and Loops in Python


### Branching with if, else and elif

```python
num1 = 38
```

```python
if num1 % 2 == 0:
    print("The given number is even-", num1)
else:
    print("The given number is odd-", num1)
```

```python
random_string = "Hello ji"
```

```python
if " " in random_string:
    print("Space is there in", random_string)
else:
    print("There is no space in ", random_string)
```

```python
today = "Wednesday"
```

```python
if today == 'Sunday':
    print("Today is the day of the sun.")
elif today == 'Monday':
    print("Today is the day of the moon.")
elif today == 'Tuesday':
    print("Today is the day of Tyr, the god of war.")
elif today == 'Wednesday':
    print("Today is the day of Odin, the supreme diety.")
elif today == 'Thursday':
    print("Today is the day of Thor, the god of thunder.")
elif today == 'Friday':
    print("Today is the day of Frigga, the goddess of beauty.")
elif today == 'Saturday':
    print("Today is the day of Saturn, the god of fun and feasting.")
```

```python
print("{} is the day of Odin, the supreme deity" .format(today))
```

```python
num2 = int(input("Enter a number:"))

if num2 % 2 == 0:
    print('{} is divisible by 2'.format(num2))
elif num2 % 3 == 0:
    print('{} is divisible by 3'.format(num2))
elif num2 % 5 == 0:
    print('{} is divisible by 5'.format(num2))
else:
    print('All checks failed!')
    print('{} is not divisible by 2, 3 or 5'.format(num2))
```

```python
num2 = int(input("Enter any number:"))
prop = "non-zero number" if (num2>0 or num2<0) else "zero"

print("{} is {}" .format(num2, prop))
```

<b> Iteration with while loops

```python
#factorial

result = 1
i = 1

while i <=10:
    result = result*i
    i+=1
    
print("Factorial of 10 is {}".format(result))
    
```

<b> Iteration with for loops

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

for day in days:
    print(day)
```

```python
# Looping over a dictionary
person = {
    'name': 'John Doe',
    'sex': 'Male',
    'age': 32,
    'married': True
}

for key in person:
    print("Key:", key, ",", "Value:", person[key])
```

```python
a_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

for i in range(len(a_list)):
    print('The value at position {} is {}.'.format(i, a_list[i]))
```

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

for day in weekdays:
    print('Today is {}'.format(day))
    if (day == 'Wednesday'):
        print("I don't work beyond Wednesday!")
        break
```

```python
persons = [{'name': 'John', 'sex': 'Male'}, {'name': 'Jane', 'sex': 'Female'}]

for person in persons:
    for key in person:
        print(key, ":", person[key])
    print(" ")
```
